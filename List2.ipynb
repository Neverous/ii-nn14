## Zadanie 3.1

```python
%%time
def BGD(function, point, step, diff):
    currentdiff = 1000000000.0
    current, gradient = function(point)
    iterations = 0
    while currentdiff >= diff:
        point -= step * gradient
        nextcurrent, gradient = function(point)
        currentdiff = numpy.abs(current - nextcurrent)
        current = nextcurrent
        iterations += 1
    
    return point, iterations

point, iterations = BGD(rosenbrock, numpy.array((0.0, 2.0)), 10e-4, 10e-10)
print iterations, '=>', point, rosenbrock(point)
```

```python
%%time
def lin_(v):
    return numpy.array((
        numpy.sum(map(lambda (c, d): 2 * v[0] - 2 * d + 2 * v[1] * c, zip(xs, ys))),
        numpy.sum(map(lambda (c, d): 2 * v[1] * c * c - 2 * d  * c + 2 * v[0] * c, zip(xs, ys))),
    ))

def lin(v):
    return numpy.sum(map(lambda (c, d): (d - v[0] - v[1] * c) ** 2, zip(xs, ys)))

def linear(v):
    return lin(v), lin_(v)

for v_0 in numpy.arange(0.0, 10.0, 0.1):
    for v_1 in numpy.arange(0.0, 10.0, 0.1):
        assert check_gradient(linear, [v_0, v_1], 10e-5)

point, iterations = BGD(linear, numpy.array((0.0, 0.0)), 10e-5, 10e-10)
print iterations, '=>', point, linear(point)

scatter(xs, ys)
x = numpy.arange(0.0, 10.0, 0.1)
plot(x, point[0] + x * point[1], 'r')
```
