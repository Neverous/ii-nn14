## Zadanie 1.4 (petla) // 11.77x

```python
%%time
TESTS = 500
def distance_squared(vec1, vec2):
    return sum(map(lambda (x, y): pow(x - y, 2), zip(vec1, vec2)))

def knneighbours(test, trainingset, classification, k):
    distances = map(lambda t: distance_squared(test, t), trainingset)
    neighbours = heapq.nsmallest(k, range(len(trainingset)), key=lambda t: distances[t])
    return collections.Counter(classification[neighbours]).most_common(1)[0][0]

iris = sklearn.datasets.load_iris()
indexes = range(len(iris.data))
tests = len(iris.data) / 3
xs = []
ys = []
for k in range(1, 20, 2):
    errors = 0
    for p in range(TESTS):
        random.shuffle(indexes)
        training = iris.data[indexes[tests:]]
        classification = iris.target[indexes[tests:]]
        for t in range(tests):
            guess = knneighbours(iris.data[indexes[t]], training, classification, k)
            errors += guess != iris.target[indexes[t]]

    xs.append(k)
    ys.append(100.0 * errors / (tests * TESTS))

fig = figure(figsize=(10, 10))
fig.suptitle('Klasyfikacja Irysow metoda K najblizszych sasiadow')
ylabel('srednia ilosc blednych odpowiedzi [%]')
xlabel('K')
plot(xs, ys, 'bx', ls='-')
```

## Zadanie 1.4 (macierze - powolne) // 1x

```python
%%time
TESTS = 500
def knneighbours(test, trainingset, classification, k):
    matrix = numpy.matrix(trainingset - test)
    distances = numpy.diag(matrix * matrix.transpose())
    return scipy.stats.mode(classification[distances.argsort()[:k]])[0][0]

iris = sklearn.datasets.load_iris()
indexes = range(len(iris.data))
tests = len(iris.data) / 3
xs = []
ys = []
for k in range(1, 20, 2):
    errors = 0
    for p in range(TESTS):
        random.shuffle(indexes)
        training = iris.data[indexes[tests:]]
        classification = iris.target[indexes[tests:]]
        for t in range(tests):
            guess = knneighbours(iris.data[indexes[t]], training, classification, k)
            errors += guess != iris.target[indexes[t]]

    xs.append(k)
    ys.append(100.0 * errors / (tests * TESTS))

fig = figure(figsize=(10, 10))
fig.suptitle('Klasyfikacja Irysow metoda K najblizszych sasiadow')
ylabel('srednia ilosc blednych odpowiedzi [%]')
xlabel('K')
plot(xs, ys, 'bx', ls='-')
```

## Zadanie 1.4 (macierze - szybsze) // 0.07x (14 razy szybsze)

```python
%%time
TESTS = 5000
ONCE = 50
def knneighbours(test, trainingset, classification, k):
    test = numpy.matrix(test)
    trainingset = numpy.matrix(trainingset)
    
    trainingset2 = numpy.diag(trainingset * trainingset.transpose()).reshape(1, len(trainingset))
    test2 = numpy.diag(test * test.transpose()).reshape(len(test), 1)
    dists2 = trainingset2 + test2 - 2 * test * trainingset.transpose()
    
    return scipy.stats.mode(classification[dists2.argsort()[:,:k]], axis=1)[0].flatten().tolist()

def knneighbours_part(test, trainingset, classification, k):
    neighbours = []
    for t in range(0, len(test), ONCE):
        neighbours.extend(knneighbours(test[t:min(len(test), t + ONCE)], trainingset, classification, k))
        
    return neighbours

iris = sklearn.datasets.load_iris()
indexes = range(len(iris.data))
tests = len(iris.data) / 3
xs = []
ys = []
for k in range(1, 20, 2):
    errors = 0
    for p in range(TESTS):
        random.shuffle(indexes)
        guess = knneighbours_part(iris.data[indexes[:tests]], iris.data[indexes[tests:]], iris.target[indexes[tests:]], k)
        errors += numpy.count_nonzero(guess != iris.target[indexes[:tests]])

    xs.append(k)
    ys.append(100.0 * errors / (tests * TESTS))

fig = figure(figsize=(10, 10))
fig.suptitle('Klasyfikacja Irysow metoda K najblizszych sasiadow')
ylabel('srednia ilosc blednych odpowiedzi [%]')
xlabel('K')
plot(xs, ys, 'bx', ls='-')
```

## Zadanie 1.4 (macierze - jeszcze troche szybsze) // 0.062x (16 razy szybsze)

```python
%%time
TESTS = 6000
ONCE = 50
def knneighbours(test, trainingset, classification, k):
    test = test.reshape(len(test), 1, len(test[0]))
    trainingset = trainingset.reshape(1, len(trainingset), len(trainingset[0]))
    diff = trainingset - test
    dists2 = numpy.sum(diff ** 2, 2)
    return scipy.stats.mode(classification[dists2.argsort()[:,:k]], axis=1)[0].flatten().tolist()

def knneighbours_part(test, trainingset, classification, k):
    neighbours = []
    for t in range(0, len(test), ONCE):
        neighbours.extend(knneighbours(test[t:min(len(test), t + ONCE)], trainingset, classification, k))
        
    return neighbours

iris = sklearn.datasets.load_iris()
indexes = range(len(iris.data))
tests = len(iris.data) / 3
xs = []
ys = []
for k in range(1, 20, 2):
    errors = 0
    for p in range(TESTS):
        random.shuffle(indexes)
        guess = knneighbours_part(iris.data[indexes[:tests]], iris.data[indexes[tests:]], iris.target[indexes[tests:]], k)
        errors += numpy.count_nonzero(guess != iris.target[indexes[:tests]])

    xs.append(k)
    ys.append(100.0 * errors / (tests * TESTS))

fig = figure(figsize=(10, 10))
fig.suptitle('Klasyfikacja Irysow metoda K najblizszych sasiadow')
ylabel('srednia ilosc blednych odpowiedzi [%]')
xlabel('K')
plot(xs, ys, 'bx', ls='-')
```
