```python
EPS = 10e-7
def numerical_gradient(function, point):
    return tuple(
        map(
            lambda (n, x):
                1.0 * (
                        function(point[:n] + [x + EPS] + point[n + 1:])
                    -   function(point[:n] + [x - EPS] + point[n + 1:])
                ) / (2.0 * EPS), enumerate(point)
        )
    )
```
