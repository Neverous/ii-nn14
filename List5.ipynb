# Zadanie 1.3

```python
def get_clusters(points, centers):
    clusters = {}
    for point in points:
        cluster = min([(i, np.linalg.norm(np.array(point)-np.array(c))) for i, c in enumerate(centers)], key=lambda x: x[1])[0]
        if cluster not in clusters:
            clusters[cluster] = []
            
        clusters[cluster].append(point)
        
    return clusters


def reevaluate_centers(centers, clusters):
    newcenters = []
    for c in sorted(clusters.keys()):
        newcenters.append(tuple(np.mean(clusters[c], axis=0)))
        
    return newcenters
    
def kmeans(points, centers):
    prev = []
    cur = [(np.random.uniform(-1.5, 6.5), np.random.uniform(-0.5, 2.5)) for _ in xrange(centers)]
    start = list(cur)
    while prev != cur:
        prev = list(cur)
        clusters = get_clusters(points, cur)
        cur = reevaluate_centers(cur, clusters)
        
    return start, cur, clusters
```

```python
COLORS = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
begin, end, clusters = kmeans(points, 8)

plt.scatter(*zip(*begin), c=COLORS, s=80, marker='*')
plt.scatter(*zip(*end), c=COLORS, s=80, marker='h')
for i, c in enumerate(sorted(clusters.keys())):
    plt.scatter(*zip(*clusters[c]), c=COLORS[i])
```

```python
def kmeans(points, centers):
    prev = []
    cur = []
    cur.append(random.choice(points))
    dists = [np.linalg.norm(np.array(point)-np.array(cur[-1])) for point in points]
    for _ in xrange(centers - 1):
        sumdists = np.sum(np.array(dists)**2)
        distribution = [dists[p]**2 / sumdists for p in xrange(len(points))]
        cur.append(points[np.random.choice(xrange(len(points)), p=distribution)])
        for p, point in enumerate(points):
            dists[p] = min(dists[p], np.linalg.norm(np.array(point)-np.array(cur[-1])))
    
    start = list(cur)
    while prev != cur:
        prev = list(cur)
        clusters = get_clusters(points, cur)
        cur = reevaluate_centers(cur, clusters)
        
    return start, cur, clusters
```
